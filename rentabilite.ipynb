```python
import matplotlib.pyplot as plt
import numpy as np

# Create the vectors X and Y
x = np.array(range(80))
xb = np.array(range(160))
y = 500 +(x/10)*1.1+1.1*(x/10)
yb = 800 + (1.3*(xb/20)) + 1.3*(xb / 20)
# Create the plot
plt.plot(x,y, label='Déneigeuse de type 1')
plt.plot(xb,yb, label='Déneigeuse de type 2')
plt.xlabel('Distance parcourue (en km)')

plt.ylabel('Coût total')
plt.legend()

# Add a title
plt.title('Comparaison des coûts des déneigeuses sur les 8 premières heures')

# Show the plot
plt.show()
```
